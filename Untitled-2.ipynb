The Newton's method for the Lagrange multiplier approach is a fast and efficient method for solving non-linear optimization problems with constraints, but it can be sensitive to the starting point, and it may not always converge to a solution. In such cases, it may be necessary to use a different optimization algorithm or to modify the Newton's method to make it more robust.



let $f(x_i)$ be the Rosenbrock function defined above. The lagrangian is:

$$ \mathcal{L}(x_{i}, \lambda) = f(x_{i}) - \lambda ( \sum_{i=1}^{n} x_{i}^{2} - r)$$

```python
#first we define the rosenbrock function with unknown parameters a and b 
def rosenbrock(x,a = 1,b = 1):
    return sum( a*(1-x[:-1])**2.0 + b*(x[1:]-x[:-1]**2.0)**2.0 )


#lagrangian
def lagrangian(x,a = 1,b = 1, lamda = 0.5, r=2): #output should be a 1 dimensional scalar value 
    return sum( a*(1-x[:-1])**2.0 + b*(x[1:]-x[:-1]**2.0)**2.0 ) - lamda * (np.sum(x**2) - r)
```




 The first n elements of the array correspond to the gradient of the Lagrangian with respect to the x variables, and the last element corresponds to the gradient with respect to the lambda variable.

 
The gradient with respect to the x variables is calculated by taking the derivative of the Lagrangian with respect to each x_i and summing the results. The gradient with respect to lambda is simply the sum of the squares of the x variables minus the value of the constraint r.



In this code, the hessian_DL function calculates the Hessian of the Lagrangian as a function of the point x and the multiplier lamda. The function returns the Hessian as a 2D numpy array, with shape (n+1, n+1), where n is the number of variables in x. The Hessian is calculated using the analytical expression for the second derivatives of the Lagrangian with respect to x and lambda. The resulting Hessian matrix is stored in the variable hessian_DL.

```python
def lagrangian_hessian_func(x, lamda = 0, r = 2, a=1 , b=1):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n - 1):
        hessian[i, i] = 2 * a + 8 * b * x[i]**2
        hessian[i, i+1] = -4 * b * x[i]
        hessian[i+1, i] = -4 * b * x[i]
        hessian[i+1, i+1] = 2 * b
    hessian = hessian - 2 * lamda * np.eye(n)
    return hessian

a = 1
b = 1
x = np.array([0.5,0.5])
hessian = lagrangian_hessian_func(x, lamda = 0.5)
print(hessian)

rank = np.linalg.matrix_rank(hessian)
if rank == hessian.shape[1]:
    print("Matrix is invertible.")
else:
    print("Matrix is not invertible.")

```


The search direction s^k is then calculated as the solution to the linear system D^2L(y^k)s^k = DL(y^k), which is obtained by multiplying the inverse of the Hessian by the gradient. The resulting search direction is returned as output. Note that the output will depend on the implementation of the hessian_DL and gradient_DL functions and may vary each time the code is run.

```python
def search_direction(y_k, lamda = 0.5, r = 2, a=1 , b=1):
    n = len(y_k)
    DL = lagrangian_gradient_func(y_k, lamda, r, a, b)
    D2L = lagrangian_hessian_func(y_k, lamda, r, a, b)
    s_k = -np.linalg.solve(D2L,  DL)
    return s_k

y = np.array([1, 2])
s = search_direction(y)
print(s)
```

```python
def newton_nextstep(y, lamda = 0.5, r=2, a = 1, b = 1, alpha=1.0):
    s = search_direction(y, lamda = 0.5, r = 2, a=1 , b=1)
    y_new = y + alpha * s
    return y_new
```

```python

def newton_function(x0, a = 1, b = 1,n=2,lamda = 0.5, r = 2, alpha = 1,  epsilon = 1e-6, max_iterations=10000 ):
    
    ''' outputs x, function_value, lamda, newton_iterations, norm_dl, constraint_deviation  '''
    x = np.zeros(n) #starting guess
    counter = 0 #iteration counter for number of newton steps 
    for i in range(max_iterations):
        gradient = lagrangian_gradient_func(x, lamda, r , a , b)
        hessian =  lagrangian_hessian_func(x, lamda , r , a , b)
        step = -np.linalg.solve(hessian, gradient) #search direction sk
        x = x + alpha * step
        if np.linalg.norm(gradient) < epsilon:
            break
        counter += 1
    
    function_value = rosenbrock(x,a ,b) #subs x back into original rosenbrock function 
    newton_iterations = counter #number of newton steps you took
    norm_dl = np.linalg.norm(gradient) #norm where we stopped 
    constraint_deviation = np.abs(np.sum(x**2) - r) #how badly are u deviating from the constraint? abs(g(x))
    return x, function_value, lamda, newton_iterations, norm_dl, constraint_deviation 

```
